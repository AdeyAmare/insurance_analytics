## Exploratory Data Analysis

```python
eda = InsuranceEDA(df_cleaned)
results = eda.run()
```

```python
# ----------------------------
# 1. Numerical Summary
# ----------------------------
print("1️⃣ Numeric Summary Statistics:")
display(results["numeric_summary"])
```

```python
# ----------------------------
# 2. Categorical Summary
# ----------------------------
print("\n2️⃣ Categorical Summary (top 10 values per column):")
for col, counts in results["categorical_summary"].items():
    print(f"\n{col}:")
    display(counts)
```

```python
# ----------------------------
# 3. Overall Loss Ratio
# ----------------------------
print("\n3️⃣ Overall Loss Ratio:")
print(results["overall_loss_ratio"])
```

```python
# ----------------------------
# 4. Grouped Loss Ratios
# ----------------------------
print("\nLoss Ratio by Province:")
print(results["grouped_loss_ratio_province"].head())

print("\nLoss Ratio by VehicleType:")
print(results["grouped_loss_ratio_vehicle"].head())

print("\nLoss Ratio by Gender:")
print(results["grouped_loss_ratio_gender"].head())
```

```python
# ----------------------------
# 5. Outliers
# ----------------------------
print("\nOutliers in key numeric columns:")
for col, out in results["outliers"].items():
    print(f"{col}: {len(out)} outliers")
```

```python
# ----------------------------
# 6. Monthly Trends
# ----------------------------
print("\nMonthly Trends (first 5 rows):")
print(results["monthly_trends"].head())
```

```python
# ----------------------------
# 7. Correlations
# ----------------------------
print("\nCorrelation Matrix:")
print(results["correlations"])
```

```python
# ----------------------------
# 8. Zip Code Aggregates
# ----------------------------
print("\nZip Code Aggregates (first 5 rows):")
print(results["zip_code_aggregates"])
```

```python
# ----------------------------
# 9. Geographic Trends
# ----------------------------
print("\nGeographic Trends by Province (first 5 rows):")
print(results["geographic_trends"]["Province"])

print("\nGeographic Trends by CoverType (first 5 rows):")
print(results["geographic_trends"]["CoverType"])

print("\nGeographic Trends by Make (first 5 rows):")
print(results["geographic_trends"]["make"])
```

```python
# ----------------------------
# 10. Vehicle Claims Summary
# ----------------------------
print("\nTop/Bottom Vehicle Claims by Make:")
print(results["vehicle_claims_summary"]["make"])
```
